프로그램 제어문 : for문 (반복문)

```python
sum = 0
for x in [1, 2, 3, 4, 5]:
  sum += x     #sum = 15
```

```python
nums = [1, 2, 3]
for x in nums:
  print(x)
```

```python
nums = range(10)     #nums = range(0, 10)
nums = range(1, 4)
sum = 0
for x in nums:
  sum = sum + x     #sum = 6
```

```python
nums = [1, 2, 3]
for x in nums:
  if x > 1: break
  print(x)
```

```python
nums = [1, 2, 3]
for x in nums:
  if x % 2 == 0: continue
  print(x)
```

프로그램 제어문 : while문 (반복문)

```python
while x <= 5:
  sum += x
  x += 1     #sum = 15
```

```python
a = 1
while a < 10:
  print(a)
  a += 1
```

```python
a = 1
while a < 10:
  print(a)
  a += 1
  if a >= 2: break
```

```python
a = 1
while a < 10:
  a += 1
  if a % 2 == 1: continue
  print(a)
```

```python
a = 1
while a < 10:
  print(a)
```

```python
a = 1
while True:
  print(a)
```

중첩 반복문 : for

```python
for x1 in [2, 3]:
  for x2 in [1, 2]:
    print(x1 * x2)
  print('뿅')
```

```python
for x1 in range(2, 10):
  for x2 in range(1, 10):
    print(x1 * x2)
```
